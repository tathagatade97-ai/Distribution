Q6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.


Ans:A uniform distribution is one in which all outcomes are equally likely. Rolling a fair die satisfies this condition.


Explanation


1.Equal probability of each outcome


A fair die has six faces (1 to 6), and each face has the same chance of appearing.


P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=1/6‚Äã


2.No bias or preference


A fair die is designed so that no number is favored, matching the key assumption of a uniform distribution.


3.Discrete uniform distribution


Since outcomes are countable and finite, rolling a die follows a discrete uniform distribution.


4.Long-run behavior


Over many rolls, each number occurs approximately the same number of times, which is exactly what a uniform distribution predicts.

Q8.How does log transformation help in stabilizing variance and making data more normally distributed?


Ans:A log transformation means replacing each data value ùë• with log(ùë•) (usually natural log or base-10 log). It is commonly used when data is right-skewed and shows non-constant variance.


1.Stabilizing Variance (Reducing Heteroscedasticity)


In many real datasets (e.g., income, sales, population), larger values show much higher variability than smaller values.


The log transformation compresses large values more than small values.


This reduces the spread of high values, making variance more constant across the range.


Example:


Raw data: 10, 100, 1000


Log data: 1, 2, 3


Large differences become manageable and more evenly spread.


2.Making Data More Normally Distributed


Right-skewed data has a long right tail.


Log transformation pulls in the right tail, reducing skewness.


The result is a distribution that is more symmetric and closer to a normal distribution.


Example:


Raw income data ‚Üí strongly right-skewed


Log(income) ‚Üí approximately bell-shaped.
